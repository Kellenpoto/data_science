## use a decorator  (no parameters)

```python
from sorters import *
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint

plt.style.use('ggplot')
%matplotlib inline
```

```python
def overplot(func):
    def in_srt_plt():
        in_sort = InsertSort()
        y = [in_sort.timed_sort(randint(100, size = i)) for i in x]
        plt.plot(x,y, label='insertion')
        func()
    return(in_srt_plt)

@overplot
def add_bu_srt_plt():
    bu_sort = BubbleSort()
    y = [bu_sort.timed_sort(randint(100, size = i)) for i in x]
    plt.plot(x,y, label='bubble') 

def add_hp_srt_plt():
    hp_sort = HeapSort()
    y = [hp_sort.timed_sort(randint(100, size = i)) for i in x]
    plt.plot(x,y, label='heap') 
    

x = np.arange(0,1000,100)

add_bu_srt_plt()
add_hp_srt_plt()
plt.legend();
```

## without decorator 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10**3,100)
in_sort = InsertSort()
bu_sort = BubbleSort()
hp_sort = HeapSort()
y1 = [in_sort.timed_sort(np.random.randint(100, size = i)) for i in x]
y2 = [bu_sort.timed_sort(np.random.randint(100, size = i)) for i in x]
y3 = [hp_sort.timed_sort(np.random.randint(100, size = i)) for i in x]

plt.plot(x,y1, label='insert')
plt.plot(x,y2, label='bubble')
plt.plot(x,y3, label = 'heap')
plt.legend();
```
